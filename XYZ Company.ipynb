# Data Preparation and Cleaning

```python
# Read the Customer Data Table form the file path
Customer = pd.read_csv(r'C:\Users\user\Desktop\Data Glacier Internship\Customer2.csv')
print(Customer)
```

```python
# Read the Transaction Data Table form the file path
Transaction = pd.read_csv(r'C:\Users\user\Desktop\Data Glacier Internship\Transaction 2.csv')
print(Transaction)
```

```python
# Read the Cab Data Table form the file path
Cab = pd.read_csv(r'C:\Users\user\Desktop\Data Glacier Internship\Cab Data 2.csv')
print(Cab)
```

```python
# Read the City Data Table form the file path
City = pd.read_csv(r'C:\Users\user\Desktop\Data Glacier Internship\City File 2.csv')
print(City)
```

```python
# Merged the Customer and Transaction Data Table together 
Customer_Transaction = pd.merge(Customer, Transaction, how='inner', on=None, left_on=None, right_on=None,
left_index=False, right_index=False, sort=True)
print(Customer_Transaction)
```

```python
# Merged the city and Cab data table
Cab_City = pd.merge(Cab, City, how='inner', on=None, left_on=None, right_on=None,
left_index=False, right_index=False, sort=True)
print(Cab_City)
```

```python
# Meged the final tables of the Cab_City and Customer_Transaction to get the final combined tables 
XYZ_Company = pd.merge(Customer_Transaction, Cab_City, how='inner', on=None, left_on=None, right_on=None,
left_index=False, right_index=False, sort=True)
print(XYZ_Company)
```

```python
# Checking for  Null Vallues
XYZ_Company.isnull().sum()
```

```python
# Remove the columns that are not required
XYZ_Company = XYZ_Company.drop(['Customer ID', 'Transaction ID', 'Payment_Mode'], axis=1)
XYZ_Company
```

```python
# Removing of Null Values from the data
XYZ_Company_Data = XYZ_Company.dropna()
XYZ_Company_Data
```

```python
# Checking for duplicates in the data
print(XYZ_Company_Data.duplicated().sum())
```

```python
# Confirming no null values in the data
print(XYZ_Company_Data.isnull().sum()) 
```

```python
# Converting the Date into Date Time format
XYZ_Company_Data['Date of Travel'] = pd.to_datetime(XYZ_Company_Data['Date of Travel'], dayfirst= True)
print(XYZ_Company_Data)
```

```python
# Creating New Column for Date, Hour and Days of the Week
XYZ_Company_Data.insert(0, 'Date', XYZ_Company_Data['Date of Travel'].dt.normalize())
XYZ_Company_Data.insert(1, 'Hour', XYZ_Company_Data['Date of Travel'].dt.hour)
XYZ_Company_Data.insert(2, 'Day of the Week', XYZ_Company_Data['Date of Travel'].dt.day_name())
```

```python
# Checking for the Distribution of the number users by the days of the week
daily_distribution = XYZ_Company_Data.groupby(['Day of the Week'], sort= False, as_index= False).mean()
daily_distribution = daily_distribution.iloc[:, [0,8]]
print(daily_distribution)
```

```python
#Relationship between Price Charged and Cost of Trip
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
x_values_for_XYZ_Company_Data = XYZ_Company_Data['Price Charged']
y_values_for_XYZ_Company_Data = XYZ_Company_Data['Cost of Trip']
XYZ_Company_Data.iloc[:,[-3,-4]]
```

```python
 # Creating an object for the linear regression class
regressor = LinearRegression()
# Reshaping the dimensions of the x_values and y_values from one_dimensional to two_dimensional
reshaped_x_values = x_values_for_XYZ_Company_Data.values.reshape(-1,1)
reshaped_y_values = y_values_for_XYZ_Company_Data.values.reshape(-1,1)
# plotting the linear regression
linear_regression = regressor.fit(reshaped_x_values, reshaped_y_values)
# creating predictions
y_pred = regressor.predict(reshaped_x_values)
# Visualizing the linear regression plot
plt.scatter(reshaped_x_values, reshaped_y_values)
plt.xlabel('Price Charged')
plt.ylabel('Cost of Trip')
plt.title('Linear Regression Curve')
plt.plot(reshaped_x_values, y_pred, color= 'red')
plt.show()

p_value = pearsonr(x_values_for_XYZ_Company_Data, y_values_for_XYZ_Company_Data)[1]
print(f'The p-value for the distribution is {p_value}')
```
